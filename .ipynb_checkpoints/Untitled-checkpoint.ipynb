```python
import numpy as np
import pandas as pd
import math
```

```python
graph = {
    'a': {'b': 85, 'c': 217, 'e': 173},
    'b': {'f': 80},
    'c': {'g': 186, 'h': 103},
    'd': dict(),
    'e': {'j': 502},
    'f': {'i': 250},
    'g': dict(),
    'h': {'d': 183, 'j': 167},
    'i': {'j': 84},
    'j': dict()
        }
```

```python
def dijkstra(graph = graph, source = 'a'):
    
    dist_matrix = pd.DataFrame(columns = sorted(graph.keys()))
    dist_matrix.loc[0] = [0] + [math.inf] * (len(dist_matrix.columns) - 1)
    
    vertex = source
    
    path = {'a': 'a'}
    
    while len(dist_matrix) <= len(dist_matrix.columns) + 1:
        
        mat_length = len(dist_matrix) - 1
        print('matrix length:', mat_length)
        
        dist_matrix.loc[vertex] = dist_matrix.iloc[mat_length]
        
        for key in graph[vertex]:
            dist_matrix.loc[vertex][key] = int(min(dist_matrix.loc[vertex][key],
                                               dist_matrix.iloc[mat_length][vertex] + graph[vertex][key]))
            
        try: # Get the next minimum distance
            vertex = dist_matrix.drop(dist_matrix.index[1:], axis = 1).loc[vertex].idxmin()
            print(vertex)
            
        except:
            return(dist_matrix)
```

```python
def dijkstra(graph = graph, source = 'a', path = True):
    
    dist_matrix = pd.DataFrame(columns = sorted(graph.keys()))
    dist_matrix.loc[0] = [0] + [math.inf] * (len(dist_matrix.columns) - 1)
    
    vertex = source
    
    path = {'a': 'a'}
    
    while len(dist_matrix) <= len(dist_matrix.columns) + 1:
        
        mat_length = len(dist_matrix) - 1
        print('matrix length:', mat_length)
        
        dist_matrix.loc[vertex] = dist_matrix.iloc[mat_length]
        
        for key in graph[vertex]:
            
            if dist_matrix.loc[vertex][key] > dist_matrix.iloc[mat_length][vertex] + graph[vertex][key]:
                dist_matrix.loc[vertex][key] = dist_matrix.iloc[mat_length][vertex] + graph[vertex][key]
                path[key] = path[vertex] + key
            
        try: # Get the next minimum distance
            vertex = dist_matrix.drop(dist_matrix.index[1:], axis = 1).loc[vertex].idxmin()
            print(vertex)
            
        except:
            if path == True:
                return(path)
            else:
                return(dist_matrix)
```

```python
# First attempt at a mapper
def mapper(graph = graph):
    
    print('# Path:\ta')
    
    for key in graph.keys():
        if key not in graph['a']:
            graph['a'][key] = math.inf
    
    for vertex in sorted(list(graph.keys())):
        for key in sorted(list(graph[vertex].keys())):
            print(vertex + key, graph[vertex][key])

mapper()
```

```python
text = """# Path:	ab
ah inf
ac 217
ad inf
ae 173
aj inf
ab 85
aa 0
af inf
ai inf
ag inf
bf 80
hd 183
hj 167
fi 250
ij 84
cg 186
ch 103
ej 502"""
```

```python
def reducer(text):
    
    a_graph = dict()
    r_graph = dict()
    
    text = text.split('\n')
    path = text[0].split('\t')[1]
    vertex = path[-1]
    
    for line in text[1:]: # First loop to define the two dictionaries
        line = line.split(' ')
        
        if line[0][0] == 'a': # First dictionary (that is going to be our final result)
            if line[1] == 'inf':
                a_graph[line[0][1]] = math.inf
            else:
                a_graph[line[0][1]] = int(line[1])
        
        else:
            r_graph[line[0]] = int(line[1]) # Second dictionary, with all 'secondary' nodes
            
            if line[0][0] == vertex:
                a_graph[line[0][1]] = min(a_graph[line[0][1]], a_graph[line[0][0]] + int(line[1]))
            
    path_graph = {key: a_graph[key] for key in a_graph if key not in path} # To get the next vertex for the algorithm
    
    try:
        print('# Path:\t%s' % path + min(path_graph, key = path_graph.get))
        
    except ValueError:
        print('# Finished')
    
    for key in a_graph:
        print('a' + key + '', a_graph[key])
        
    for key in r_graph:
        print(key + '', r_graph[key])

reducer(text = text)
```

```python
import scipy.sparse
import os
import pandas as pd
import numpy as np
```

```python
def dijkstra_check(text):
    
    sparse_matrix = np.array(text).transpose()
    
    dimension = max(max(sparse_matrix[0]), max(sparse_matrix[1]))

    matrix = np.zeros((dimension, dimension))

    for ind1, ind2, weight in zip(sparse_matrix[0], sparse_matrix[1], sparse_matrix[2]):
        ind1 -= 1
        ind2 -= 1
        matrix[ind1][ind2] = weight
        matrix[ind2][ind1] = weight
     
    return(pd.DataFrame(scipy.sparse.csgraph.dijkstra(matrix),
                        columns = range(1, dimension + 1),
                        index = range(1, dimension + 1)))

text = pd.read_csv('sparse_matrix_3.txt', sep = ";", names = ['a','b','c'], dtype = int)

dijkstra_check(text)
```
